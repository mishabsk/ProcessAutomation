```python
# Example: Visualizing Attention Patterns (Conceptual)

import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

def visualize_attention(attention_matrix, tokens):
    """
    Visualize attention weights as a heatmap.

    Args:
        attention_matrix: 2D numpy array of attention scores
        tokens: List of token strings
    """
    plt.figure(figsize=(10, 8))
    sns.heatmap(attention_matrix,
                xticklabels=tokens,
                yticklabels=tokens,
                cmap='YlOrRd',
                annot=True,
                fmt='.2f',
                cbar_kws={'label': 'Attention Weight'})
    plt.title('Self-Attention Visualization', fontsize=14, fontweight='bold')
    plt.xlabel('Keys (attending to)', fontsize=12)
    plt.ylabel('Queries (attending from)', fontsize=12)
    plt.tight_layout()
    plt.show()

# Example: Simulated attention pattern
tokens = ['The', 'cat', 'sat', 'on', 'the', 'mat']
n_tokens = len(tokens)

# Create synthetic attention pattern (diagonal + some cross-attention)
attention = np.eye(n_tokens) * 0.6  # Strong self-attention
attention += np.random.rand(n_tokens, n_tokens) * 0.2  # Random attention
attention[1, 5] = 0.8  # "cat" attends to "mat"
attention[5, 1] = 0.7  # "mat" attends to "cat"

# Normalize rows to sum to 1 (softmax-like)
attention = attention / attention.sum(axis=1, keepdims=True)

visualize_attention(attention, tokens)

print("\n💡 Interpretation:")
print("- Diagonal values show self-attention (each token attends to itself)")
print("- Off-diagonal values show cross-attention between tokens")
print("- Darker colors indicate stronger attention relationships")
```

```python
# Text Generation Example using Hugging Face Transformers

from transformers import pipeline, set_seed
import warnings
warnings.filterwarnings('ignore')

# Set seed for reproducibility
set_seed(42)

def demonstrate_text_generation():
    """
    Demonstrate text generation with different prompts and parameters.
    """
    print("🤖 Loading text generation model...")

    # Initialize the text generation pipeline with a smaller model
    # Using GPT-2 as it's widely available and doesn't require authentication
    generator = pipeline('text-generation', model='gpt2', device=-1)  # CPU

    print("✅ Model loaded successfully!\n")

    # Example 1: Creative writing
    print("="*70)
    print("📝 Example 1: Creative Story Generation")
    print("="*70)

    prompt1 = "In a world where artificial intelligence has become sentient,"
    print(f"\nPrompt: {prompt1}")
    print("\nGenerated Text:")

    result1 = generator(
        prompt1,
        max_length=100,
        num_return_sequences=1,
        temperature=0.8,  # Controls randomness (higher = more creative)
        top_p=0.9,        # Nucleus sampling
        do_sample=True
    )
    print(result1[0]['generated_text'])

    # Example 2: Technical writing
    print("\n" + "="*70)
    print("🔬 Example 2: Technical Explanation")
    print("="*70)

    prompt2 = "The fundamental difference between machine learning and deep learning is that"
    print(f"\nPrompt: {prompt2}")
    print("\nGenerated Text:")

    result2 = generator(
        prompt2,
        max_length=80,
        num_return_sequences=1,
        temperature=0.5,  # Lower temperature for more focused output
        do_sample=True
    )
    print(result2[0]['generated_text'])

    # Example 3: Multiple completions
    print("\n" + "="*70)
    print("🎲 Example 3: Multiple Completion Variations")
    print("="*70)

    prompt3 = "The future of education will be"
    print(f"\nPrompt: {prompt3}")
    print("\nGenerating 3 different completions...\n")

    results3 = generator(
        prompt3,
        max_length=60,
        num_return_sequences=3,
        temperature=0.9,
        do_sample=True
    )

    for i, result in enumerate(results3, 1):
        print(f"Version {i}:")
        print(result['generated_text'])
        print()

    print("\n💡 Key Parameters Explained:")
    print("   • max_length: Maximum number of tokens to generate")
    print("   • temperature: Controls randomness (0.0-2.0, higher = more creative)")
    print("   • top_p: Nucleus sampling threshold (keeps top tokens summing to p)")
    print("   • num_return_sequences: Number of different completions to generate")

# Uncomment to run (requires transformers and torch installed)
demonstrate_text_generation()
```

```python
# Machine Translation Example

def demonstrate_translation():
    """
    Demonstrate multilingual translation capabilities.
    """
    from transformers import pipeline

    # Example using a multilingual model
    translator_fr = pipeline(
        "translation",
        model="Helsinki-NLP/opus-mt-en-fr",  # English to French
        device=-1
    )

    translator_de = pipeline(
        "translation",
        model="Helsinki-NLP/opus-mt-en-de",  # English to German
        device=-1
    )

    translator_es = pipeline(
        "translation",
        model="Helsinki-NLP/opus-mt-en-es",  # English to Spanish
        device=-1
    )

    # Sample sentences
    sentences = [
        "Large language models are transforming artificial intelligence.",
        "Machine translation has improved significantly with deep learning.",
        "The future of natural language processing is very exciting."
    ]

    print("="*70)
    print("🌍 MULTILINGUAL TRANSLATION DEMONSTRATION")
    print("="*70)

    for i, sentence in enumerate(sentences, 1):
        print(f"\n--- Example {i} ---")
        print(f"🇬🇧 English:  {sentence}")

        # Translate to French
        fr_result = translator_fr(sentence)
        print(f"🇫🇷 French:   {fr_result[0]['translation_text']}")

    print("\n" + "="*70)
    print("💡 Translation Quality Factors:")
    print("   • Model size and training data")
    print("   • Language pair similarity")
    print("   • Domain-specific terminology")
    print("   • Context and ambiguity resolution")

demonstrate_translation()
```

```python
# Sentiment Analysis Example

def demonstrate_sentiment_analysis():
    """
    Perform sentiment analysis on various text samples.
    """
    from transformers import pipeline
    import pandas as pd

    print("🤖 Loading sentiment analysis model...")

    # Using DistilBERT model fine-tuned for sentiment analysis
    classifier = pipeline(
        "sentiment-analysis",
        model="distilbert-base-uncased-finetuned-sst-2-english",
        device=-1
    )

    print("✅ Model loaded!\n")

    # Test samples with varying sentiments
    reviews = [
        "This product is absolutely amazing! Best purchase ever.",
        "I'm very disappointed with the quality. Would not recommend.",
        "It's okay, nothing special but gets the job done.",
        "Terrible experience. Complete waste of money.",
        "Exceeded my expectations! Highly recommend to everyone.",
        "The product is fine but customer service was unhelpful.",
        "Love it! Will definitely buy again.",
        "Not what I expected based on the description."
    ]

    print("="*70)
    print("😊 SENTIMENT ANALYSIS RESULTS")
    print("="*70 + "\n")

    results = []

    for i, review in enumerate(reviews, 1):
        result = classifier(review)[0]
        sentiment = result['label']
        confidence = result['score']

        # Emoji representation
        emoji = "😊" if sentiment == "POSITIVE" else "😞"

        results.append({
            'Review': review[:50] + '...' if len(review) > 50 else review,
            'Sentiment': sentiment,
            'Confidence': f"{confidence:.2%}"
        })

        print(f"{emoji} Review {i}: {review}")
        print(f"   Sentiment: {sentiment} (Confidence: {confidence:.2%})\n")

    # Create summary DataFrame
    df = pd.DataFrame(results)

    print("="*70)
    print("📊 SUMMARY")
    print("="*70)
    print(df.to_string(index=False))

    # Statistics
    positive_count = sum(1 for r in results if r['Sentiment'] == 'POSITIVE')
    negative_count = len(results) - positive_count

    print(f"\n📈 Statistics:")
    print(f"   Positive: {positive_count} ({positive_count/len(results):.1%})")
    print(f"   Negative: {negative_count} ({negative_count/len(results):.1%})")

    print("\n💡 Applications:")
    print("   • Customer feedback analysis")
    print("   • Social media monitoring")
    print("   • Product review aggregation")
    print("   • Brand reputation tracking")

# Uncomment to run
demonstrate_sentiment_analysis()
```

```python
# Zero-Shot Text Classification Example (NO EXAMPLES)

def demonstrate_text_classification():
    """
    Demonstrate zero-shot text classification with custom categories.
    """
    from transformers import pipeline

    print("🤖 Loading zero-shot classification model...")

    # Using BART-based zero-shot classifier
    classifier = pipeline(
        "zero-shot-classification",
        model="facebook/bart-large-mnli",
        device=-1
    )

    print("✅ Model loaded!\n")

    # Define categories
    categories = [
        "Technology",
        "Sports",
        "Politics",
        "Entertainment",
        "Health",
        "Business"
    ]

    # Sample texts
    texts = [
        "Apple announces new AI-powered features in latest iPhone release.",
        "The Lakers won the championship after an intense final game.",
        "New healthcare policy aims to expand coverage to millions.",
        "Box office records broken by latest superhero movie.",
        "Stock markets rally as tech companies report strong earnings.",
        "Researchers develop new treatment for rare disease."
    ]

    print("="*70)
    print("🏷️  ZERO-SHOT TEXT CLASSIFICATION")
    print("="*70)
    print(f"\nAvailable Categories: {', '.join(categories)}\n")

    for i, text in enumerate(texts, 1):
        result = classifier(text, candidate_labels=categories)

        print(f"--- Text {i} ---")
        print(f"📄 {text}")
        print(f"\n🎯 Top Predictions:")

        # Show top 3 categories
        for j in range(min(3, len(result['labels']))):
            label = result['labels'][j]
            score = result['scores'][j]
            bar = "█" * int(score * 20)
            print(f"   {j+1}. {label:<15} {score:.2%} {bar}")

        print()

    print("💡 Zero-Shot Classification Benefits:")
    print("   • No training data needed for new categories")
    print("   • Flexible - add/remove categories easily")
    print("   • Cost-effective for dynamic classification needs")
    print("   • Works across different domains")

# Uncomment to run
demonstrate_text_classification()
```

```python
# BLEU Score Calculation Example

def calculate_bleu_example():
    """
    Demonstrate BLEU score calculation for machine translation evaluation.
    """
    import nltk
    from nltk.translate.bleu_score import sentence_bleu, SmoothingFunction

    # Uncomment if needed
    # nltk.download('punkt')

    print("="*70)
    print("📊 BLEU SCORE EVALUATION")
    print("="*70 + "\n")

    # Reference translations (ground truth)
    reference1 = "The cat is sitting on the mat".split()
    reference2 = "A cat sits on a mat".split()
    references = [reference1, reference2]

    # Candidate translations (model outputs)
    candidates = [
        "The cat is sitting on the mat".split(),  # Perfect match
        "The cat sits on the mat".split(),        # Close match
        "A feline is on the mat".split(),         # Paraphrase
        "The dog is running in the park".split()  # Poor match
    ]

    descriptions = [
        "Perfect match (identical to reference 1)",
        "Close match (slight variation)",
        "Paraphrase (different words, similar meaning)",
        "Poor match (completely different meaning)"
    ]

    smooth_func = SmoothingFunction().method4

    print("Reference Translations:")
    print(f"  1. {' '.join(reference1)}")
    print(f"  2. {' '.join(reference2)}")
    print("\n" + "-"*70 + "\n")

    for i, (candidate, desc) in enumerate(zip(candidates, descriptions), 1):
        score = sentence_bleu(references, candidate, smoothing_function=smooth_func)

        print(f"Candidate {i}: {' '.join(candidate)}")
        print(f"Description: {desc}")
        print(f"BLEU Score: {score:.4f}")

        # Visual representation
        bar_length = int(score * 50)
        bar = "█" * bar_length + "░" * (50 - bar_length)
        print(f"Score: [{bar}] {score:.1%}")
        print()

    print("="*70)
    print("💡 BLEU Score Interpretation:")
    print("   • 1.0: Perfect match with reference")
    print("   • 0.6-1.0: Excellent translation")
    print("   • 0.4-0.6: Good translation")
    print("   • 0.2-0.4: Understandable but needs improvement")
    print("   • 0.0-0.2: Poor translation")
    print("\n   Note: BLEU has limitations - it doesn't capture semantic")
    print("   equivalence well and can favor word-for-word translations.")

# Uncomment to run
calculate_bleu_example()
```

```python
# ROUGE Score Calculation Example

def calculate_rouge_example():
    """
    Demonstrate ROUGE score calculation for summarization evaluation.
    """
    from rouge_score import rouge_scorer

    print("="*70)
    print("📊 ROUGE SCORE EVALUATION")
    print("="*70 + "\n")

    # Reference summary (ground truth)
    reference = """Large language models are transforming artificial intelligence
    by enabling machines to understand and generate human language with unprecedented
    accuracy and fluency."""

    # Candidate summaries
    candidates = [
        "Large language models are transforming AI by enabling machines to understand and generate human language.",
        "LLMs help computers process language better than ever before.",
        "Artificial intelligence has been improved by language models.",
        "Neural networks can now generate coherent text using transformers."
    ]

    descriptions = [
        "Close paraphrase of reference",
        "Concise but captures main idea",
        "Too abstract, loses specifics",
        "Different focus, mentions specifics not in reference"
    ]

    # Initialize ROUGE scorer
    scorer = rouge_scorer.RougeScorer(['rouge1', 'rouge2', 'rougeL'], use_stemmer=True)

    print("Reference Summary:")
    print(f"  {reference}")
    print("\n" + "-"*70 + "\n")

    for i, (candidate, desc) in enumerate(zip(candidates, descriptions), 1):
        scores = scorer.score(reference, candidate)

        print(f"Candidate {i}: {candidate}")
        print(f"Description: {desc}")
        print(f"\nROUGE Scores:")

        for metric, score in scores.items():
            print(f"  {metric.upper()}:")
            print(f"    Precision: {score.precision:.4f}")
            print(f"    Recall:    {score.recall:.4f}")
            print(f"    F1-score:  {score.fmeasure:.4f}")
        print()

    print("="*70)
    print("💡 ROUGE Metrics Explained:")
    print("   • ROUGE-1: Unigram (single word) overlap")
    print("   • ROUGE-2: Bigram (two-word phrase) overlap")
    print("   • ROUGE-L: Longest common subsequence")
    print("\n   • Precision: % of candidate words in reference")
    print("   • Recall: % of reference words in candidate")
    print("   • F1: Harmonic mean of precision and recall")

# Uncomment to run
```
